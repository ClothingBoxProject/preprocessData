```python
import pandas as pd
import re
```

```python
df = pd.read_csv("res_woowarhanclean.csv", encoding="utf-8")
print("원본 개수:", len(df))
df.head()
```

```python
# 정규식: 공백 있어도/없어도 매칭
mask = df["title"].str.contains(r"버리는", regex=True, na=False)

kept = df[mask].copy()
dropped = df[~mask].copy()

print("필터 후 개수:", len(kept))
print("버린 개수:", len(dropped))

kept.head()
```

```python
def extract_item_name(title):
    if pd.isna(title):
        return None
    
    before = re.split(r"버리는\s*(?:방법|법)", str(title), maxsplit=1)[0].strip()
    if not before:
        return None
    before = re.sub(r"[^\w가-힣\s]", "", before)
    return before.strip()
```

```python
kept["item_name"] = kept["title"].apply(extract_item_name)
kept[["title", "item_name"]].head(10)
```

```python
kept.to_csv("filtered_items.csv", index=False, encoding="utf-8-sig")
```

```python
def normalize_text(s):
    if pd.isna(s):
        return ""
    s = str(s)
    s = s.replace("\r\n", "\n").replace("\r", "\n")
    s = re.sub(r"[ \t]+", " ", s)
    s = re.sub(r"\n{3,}", "\n\n", s)
    return s.strip()
```

```python
def parse_sections(full_text: str):
    text = normalize_text(full_text)
    if not text:
        return None, None, None

    lines = [ln.strip() for ln in text.split("\n") if ln.strip()]

    # 1) one_line extract
    one_line = None
    one_line_idx = None
    for i, ln in enumerate(lines):
        if "1줄요약" in ln:
            m = re.search(r"1줄요약\s*[:：]\s*(.+)$", ln)
            if m:
                one_line = m.group(1).strip()
                one_line_idx = i
            else:
                if i + 1 < len(lines):
                    one_line = lines[i + 1].strip()
                    one_line_idx = i
            break

    # 2) caution split
    caution_start = None
    for i, ln in enumerate(lines):
        if re.search(r"알아두면 좋은\s*점", ln):
            caution_start = i
            break

    # 3) how_to_throw split : (one_line 라인 이후) ~ (알아두면 좋은 점 전)
    start = 0
    if one_line_idx is not None:
        start = one_line_idx + 1

    end = len(lines)
    if caution_start is not None:
        end = caution_start

    how_lines = lines[start:end]

    # how_to_throw에서 "1줄요약" 라인 자체가 끼어들면 제거
    how_lines = [ln for ln in how_lines if "1줄요약" not in ln]
    how_to_throw = "\n".join(how_lines).strip() if how_lines else None

    # caution
    caution = None
    if caution_start is not None:
        caution_lines = lines[caution_start+1:]  # 헤더 줄 제외
        caution = "\n".join(caution_lines).strip() if caution_lines else None
    return one_line, how_to_throw, caution
```

```python
modified = pd.read_excel("filtered_final.xlsx")
```

```python
# text 컬럼 정규화(선택)
modified["text"] = modified["text"].apply(normalize_text)
parsed = modified["text"].apply(parse_sections)

modified["one_line"] = parsed.apply(lambda x: x[0])
modified["how_to_throw"] = parsed.apply(lambda x: x[1])
modified["caution"] = parsed.apply(lambda x: x[2])
modified[["title", "one_line", "how_to_throw", "caution"]].head(3)
```

```python
import argparse
from pathlib import Path
import pandas as pd
import os
```

```python
IMG_ROOT = Path("./unsplash_photos/img")
NO_IMG = Path("./unsplash_photos/no_img/unknown.jpg")
IMG_EXTS = {".jpg", ".jpeg", ".png", ".webp", ".avif"}


def pick_first_image(folder: Path):
    if not folder.exists() or not folder.is_dir():
        return None
    files = sorted([p for p in folder.iterdir() if p.suffix.lower() in IMG_EXTS and p.is_file()])
    return files[0] if files else None


def main():
    ap = argparse.ArgumentParser()
    ap.add_argument("--input", default="./data/filtered_postprocessed.xlsx")
    ap.add_argument("--col", default="item_name")
    ap.add_argument("--out", default="./data/clothing_type.csv")
    ap.add_argument("--url-col", default="url")
    ap.add_argument("--as-file-uri", action="store_true")
    args, _unknown = ap.parse_known_args()

    in_path = Path(args.input)
    out_path = Path(args.out)

    if in_path.suffix.lower() in [".xlsx", ".xls"]:
        df = pd.read_excel(in_path)
    elif in_path.suffix.lower() == ".csv":
        df = pd.read_csv(in_path)
    else:
        raise ValueError("지원: .xlsx 또는 .csv")

    if args.col not in df.columns:
        raise KeyError(f"'{args.col}' 컬럼이 없음. 현재 컬럼: {list(df.columns)}")

    PROJECT_ROOT = Path.cwd()

    def to_url(p: Path) -> str:
        p = p.resolve()
        if args.as_file_uri:
            return p.as_uri()

        rel = os.path.relpath(str(p), start=str(PROJECT_ROOT))
        return Path(rel).as_posix()

    urls = []
    for item in df[args.col].fillna("").astype(str):
        item = item.strip()

        img_path = pick_first_image(IMG_ROOT / item) if item else None
        if img_path is None:
            img_path = NO_IMG

        urls.append(to_url(img_path))

    df[args.url_col] = urls

    out_path.parent.mkdir(parents=True, exist_ok=True)

    if out_path.suffix.lower() in [".xlsx", ".xls"]:
        df.to_excel(out_path, index=False)
    elif out_path.suffix.lower() == ".csv":
        df.to_csv(out_path, index=False, encoding="utf-8-sig")
    else:
        raise ValueError("출력은 .xlsx 또는 .csv만 지원")

    print(f"Done. Saved: {out_path}")
    print(f"Example url: {df[args.url_col].iloc[0]}")


if __name__ == "__main__":
    main()
```
